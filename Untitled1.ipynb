```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics as stat
```

```python
from sklearn.datasets import fetch_california_housing
california = fetch_california_housing()
```

```python
dataset = pd.DataFrame(california.data, columns=california.feature_names)
dataset.head()
```

```python
print(california.DESCR)
```

```python
dataset.to_csv('california.csv')
```

```python
pd.read_csv('/content/drive/MyDrive/BigData_Ever/california.csv')
dataset.head()
print(dataset.columns)
print("cantidad de variables", len(dataset.columns))
variables=dataset.columns

for i in range(len(variables)):
  data=dataset[variables[i]]
  #print(variables[i])
  contar=len(data)
  promedio=np.sum(data)/contar
  mediana=np.median(data)
  moda=stat.mode(data)
  desviacion=np.std(data)
 #print(suma)
  #print(variables[i],contar,promedio,mediana,moda,desviacion)
  print(variables[i],promedio,mediana,moda)
```

```python
# para obtener los estadisticos básicos
descripcion=dataset.describe()
print(descripcion)
```

```python
fig=plt.figure(figsize=(20,20))
k=0
le=len(variables)
for j in range(le):
  for i in range(le):
    k=k+1
    ax=plt.subplot(le, le, k)
    ax.scatter(dataset[variables[j]],dataset[variables[i]])
    ax.set_xlabel(variables[j])
    ax.set_ylabel(variables[i])
    fig.tight_layout()
plt.show()
```
